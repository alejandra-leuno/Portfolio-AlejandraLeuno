# MODELOS

```python
# Importación de librerías
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import train_test_split, KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

```python
# División del dataset en un conjunto de entrenamiento (90%) y un conjunto de testeo (10%)
X = data.drop('charges', axis = 1)
y = data['charges']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

```python
print(data.shape)
```

```python
print("Tamaño del conjunto de entrenamiento:", X_train.shape)
print("Tamaño del conjunto de prueba:", X_test.shape)
```

# Regresión Lineal Ridge Cross Validation

```python
# Se divide el conjunto de entrenamiento en 4 folds.

# Se definen los valores de Alpha a probar
alphas = np.arange(0.01, 0.41, 0.01)

# Modelo Ridge
reg_ridgeCV =RidgeCV(alphas=alphas, cv=4)
reg_ridgeCV.fit(X_train, y_train)
```

```python
# Predicciones en el conjunto de testeo para Ridge
y_test_pred_ridgeCV = reg_ridgeCV.predict(X_test)

# Cálculo del error cuadrático medio en el conjunto de testeo
mse_ridgeCV = mean_squared_error(y_test, y_test_pred_ridgeCV)

print("Mejor alpha encontrado:", reg_ridgeCV.alpha_)
print("Error cuadrático medio (RidgeCV):", mse_ridgeCV)
```

```python
mse_alphas = []
alphas = np.arange(0.01, 0.41, 0.01)
for i in alphas:
  # Creación y entrenamiento del modelo de regresión con Ridge
  reg_ridgeCV = RidgeCV(alphas=i)
  reg_ridgeCV.fit(X_train, y_train)

  # Predicciones en el conjunto de predicción
  y_val_pred_ridgeCV = reg_ridgeCV.predict(X_test)

  # Cálculo del error cuadrático medio en el conjunto de testeo
  mse_ridgeCV = r2_score(y_test, y_val_pred_ridgeCV)
  mse_alphas.append(mse_ridgeCV)

  print("Error cuadrático medio (Ridge) con alpha = {} es {}".format(round(i, 2), round(mse_ridgeCV, 2)))
```

# Arbol de decisión

```python
# Módulos específicos para construcción y visualización del modelo
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
# Construcción del árbol de decisión de profundidad 5
dtr = DecisionTreeRegressor(max_depth = 5, random_state = 42)
```

```python
# Entrenamiento del modelo con el método fit
dtr.fit(X_train, y_train)
```

```python
# Score del modelo en los datos de entrenamiento
score_in_train = dtr.score(X_train, y_train)

print('Score en el set de entrenamiento: {}'.format(score_in_train))
```

```python
# Score del modelo en los datos de testeo
score_in_test = dtr.score(X_test, y_test)

print('Score en el set de testeo: {}'.format(score_in_test))
```

# Random Forests

```python
# Módulo especifíco para el modelo
from sklearn.ensemble import RandomForestRegressor
```

```python
# Construcción del modelo con 100 estimadores
# Por default, cada árbol individual es entrenado hasta máxima profundidad
rfr = RandomForestRegressor(n_estimators=100)
```

```python
# Entrenamiento del modelo
rfr.fit(X_train, y_train)
```

```python
# Score en el set de entrenamiento
score_in_train = rfr.score(X_train, y_train)
print('Score en el set de train: {}'.format(score_in_train))
```

```python
# Score en el set de testeo
score_in_test = rfr.score(X_test, y_test)
print('Score en el set de testeo: {}'.format(score_in_test))
```
